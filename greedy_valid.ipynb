```python
#刪除孤單的點
lonely_node = []
for NI in G1.Nodes():
    if NI.GetOutDeg()==0:
        lonely_node.append(NI.GetId())
        G1.DelNode(NI.GetId())

print("lonely_node = ",lonely_node)
```

```python
S = 0 #初始感染點
AllnodeID = [] #全部的node
InfectednodeID = [S]
vulnerablenodeID = []
vaccinatednodeID = [] #vaccinated的node
# inf_plus_vul_minus_zero = []
for NI in G1.Nodes():
    AllnodeID.append(NI.GetId())
    if(NI.GetId()!=S):
        vulnerablenodeID.append(NI.GetId())
    print("node: %d" % NI.GetId())
diam = G1.GetBfsFullDiam(len(AllnodeID), False)#直徑
```

```python
print("time = 0")
print("infect_nodeID",InfectednodeID)
print("vulnerable_nodeID=",vulnerablenodeID)
print("vaccinated_nodeID",vaccinatednodeID)
strategy = []
 #注射的點 +被保護的點
r = 1
while len(vulnerablenodeID)!=0:
    saved = [0]*len(vulnerablenodeID) #保護node數量
    temp_saved = []
    for i in range(len(vulnerablenodeID)):
        count = 0 #是否有被保護
        temp_saved.append([vulnerablenodeID[i]]) #注射的點
        for j in range(len(vulnerablenodeID)): #每一個點計算是否有被注射的點保護
            if(r<=(G1.GetShortPath(S,vulnerablenodeID[j])-G1.GetShortPath(vulnerablenodeID[i],vulnerablenodeID[j]))):
                count = count+1
                temp_saved[i].append(vulnerablenodeID[j]) #被保護的點
        saved[i] = count
        
    print("time=",r)    
    print("vaccinated_node=",temp_saved[saved.index(max(saved))][0]) #選擇可以保護最多點的點去注射 
    strategy.append(temp_saved[saved.index(max(saved))][0])
    for i in range(1,len(temp_saved[saved.index(max(saved))])):
        print("saved_node=",temp_saved[saved.index(max(saved))][i])
        G1.DelNode(temp_saved[saved.index(max(saved))][i]) #圖中刪除已被保護的點
        vaccinatednodeID.append(temp_saved[saved.index(max(saved))][i]) 
        vulnerablenodeID.remove(temp_saved[saved.index(max(saved))][i])
        
    vulnerablenodeId_temp = vulnerablenodeID.copy()
    for i in range(len(vulnerablenodeId_temp)): #病毒蔓延一層
        if G1.GetShortPath(S, vulnerablenodeId_temp[i])==r:
            InfectednodeID.append(vulnerablenodeId_temp[i])
            vulnerablenodeID.remove(vulnerablenodeId_temp[i])
            
    print("infect_nodeID",InfectednodeID)
    print("vulnerable_nodeID=",vulnerablenodeID)
    print("vaccinated_nodeID",vaccinatednodeID)
    
    r = r+1
        
```

```python
end_time = time.time()
```

```python
print("最佳策略=",strategy)
print("保護個數=",len(vaccinatednodeID))
print("被保護的點=",vaccinatednodeID)
print("計算時間=",end_time-start_time,"sec")
```
